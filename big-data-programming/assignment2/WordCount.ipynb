```python
import findspark
findspark.init()

import pyspark

# test.txt file
file=open("test.txt","r+")
wordcount={}

for word in file.read().split():
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
        
for k,v in wordcount.items():
    print (k, v)
```
